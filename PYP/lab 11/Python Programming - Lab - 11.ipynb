### 02) WAP to to Plot List random of X, Y Coordinates in Matplotlib.

```python
import random
x = []
y = []
for i in range(1,11) :
    a = random.randint(1,10)
    b = random.randint(1,10)
    x.append(a)
    y.append(b)
print(x)
print(y)
plt.plot(x,y)
plt.show()
```

### 03) WAP to demonstrate the use of Bar chart.

```python
cxMarks= [8,9,6,3,5]
cyMarks= [5,8,9,6,3]
l = ['1st','2nd','3rd','4th','5th']
c = ['b','g','r','c','m']
plt.title("Marks")
plt.bar(cxMarks,cyMarks,color=c,label=l)
for i in range(len(cyMarks)) :
    plt.text(cxMarks[i],cyMarks[i],cyMarks[i])
plt.show()
```

### 04) WAP to demonstrate the use of Histogram.

```python
import random
cpi = []
for i in range(100) :
    a = random.randint(1,10)
    cpi.append(a)
plt.hist(cpi,bins=10,histtype='stepfilled',align='mid',label='CPI Hist')
plt.legend()
plt.show()
```
